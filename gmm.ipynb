```python
import sklearn 
import sklearn.datasets
import numpy as np
import sklearn.mixture
import matplotlib.pyplot as plt
import scipy
```

```python
co_type = "full"
X, y = sklearn.datasets.make_blobs(n_samples=200, centers=4, n_features=2, cluster_std=1, random_state=0)
bgm = sklearn.mixture.BayesianGaussianMixture(covariance_type = co_type, n_components=5, random_state=42, weight_concentration_prior=0.01, init_params='k-means++').fit(X)

Z = bgm.predict(X)

colors = ["red", "blue", "green", "orange", "darkblue", "purple", "black"]
for k, col in enumerate(colors):
    cluster_data = Z == k
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1], c=col, marker=".", s=20)
plt.scatter(bgm.means_[:,0], bgm.means_[:,1], c="black", marker='*')
plt.title("GMM model with 5 specified clusters")
plt.show()
```
